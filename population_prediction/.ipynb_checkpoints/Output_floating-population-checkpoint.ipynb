```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
STAMP = 1 ; LAG = 10 ; STEP = 1 
gt = pd.read_csv(f'./output/predictions/stamp{STAMP}_lag{LAG}_step{STEP}_gt.csv')
pred = pd.read_csv(f'./output/predictions/stamp{STAMP}_lag{LAG}_step{STEP}_pred.csv')
```

```python
plt.figure(figsize=(15,5))
plt.plot(gt['col_93'], color='black', label='True')
plt.plot(pred['col_93'], color='red', label='Prediction')
plt.title('Prediction of location 93', fontsize=20)
plt.legend(loc='best')
plt.ylim([0,200])
plt.show()

# plt.figure(figsize=(20,7))
# plt.plot(df_37["datetime"][:200], df_37["tgnet_pred"][:200], color='black', label='TGNET')
# plt.plot(df_37["datetime"][:200], df_37["pickup_true"][:200], color='r', label='True')
# plt.xticks([])
# plt.title('Comparison TGNET Prediction and True')
```

```python
gt_np = np.array(gt).reshape([len(gt),10,20])
pred_np = np.array(pred).reshape([len(pred),10,20])
```

```python
idx = 0 ; c = 20 ; vmax = 200
plt.figure(figsize=(10,20))
plt.imshow(gt_np[idx+18][:][:] + c, cmap='gray', vmin=0, vmax=vmax) # 00:00
plt.title('2022-04-10 12:00', fontsize=20)
plt.show()
```

```python
m = 10
from datetime import *
target = datetime(2022,4,1,0,0) + timedelta(minutes=30*(m+1))
str(target)
```

```python
idx = 0 ; c = 20 ; vmax = 200 ; m = 10

for i in range(1,48,6):
    fig = plt.figure(figsize=(12,6))
    target = datetime(2022,4,1,0,0) + timedelta(minutes=30*(m+i+1))
    # True
    ax1 = fig.add_subplot(1,2,1)
    ax1.imshow(gt_np[i][:][:] + c, cmap='gray', vmin=0, vmax=vmax)
    ax1.set_title(f'True   {str(target)}', fontsize=15)
    
    # Prediction
    ax2 = fig.add_subplot(1,2,2)
    ax2.imshow(pred_np[i][:][:] + c, cmap='gray', vmin=0, vmax=vmax)
    ax2.set_title(f'Prediction  {str(target)}', fontsize=15)
    
    plt.show()
    
```
