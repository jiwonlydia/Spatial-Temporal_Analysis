## Single-step Forecast

```python
STAMP = 1 ; LAG = 10 ; STEP = 48 
gt30 = pd.read_csv(f'./output/predictions/stamp{STAMP}_lag{LAG}_step{STEP}_gt.csv')
pred30 = pd.read_csv(f'./output/predictions/stamp{STAMP}_lag{LAG}_step{STEP}_pred.csv')
```

```python
STAMP = 48 ; LAG = 10 ; STEP = 1 
gt24 = pd.read_csv(f'./output/predictions/stamp{STAMP}_lag{LAG}_step{STEP}_gt.csv')
pred24 = pd.read_csv(f'./output/predictions/stamp{STAMP}_lag{LAG}_step{STEP}_pred.csv')
```

```python
gt_np24 = np.array(gt24).reshape([len(gt24),10,20])
pred_np24 = np.array(pred24).reshape([len(pred24),10,20])
```

```python
c = 20 ; vmax = 200 ; m = 10

for i in range(0,72,6):
    fig = plt.figure(figsize=(12,6))
    target = datetime(2022,4,1,0,0) + timedelta(minutes=30*(48*m+i))
    # True
    ax1 = fig.add_subplot(1,2,1)
    ax1.imshow(gt_np24[i][:][:] + c, cmap='gray', vmin=0, vmax=vmax)
    ax1.set_title(f'True   {str(target)}', fontsize=15)
    
    # Prediction
    ax2 = fig.add_subplot(1,2,2)
    ax2.imshow(pred_np24[i][:][:] + c, cmap='gray', vmin=0, vmax=vmax)
    ax2.set_title(f'Prediction  {str(target)}', fontsize=15)
    
    plt.show()
    
```

```python
i = 36
target = datetime(2022,4,1,0,0) + timedelta(minutes=30*(48*m+i))
plt.figure(figsize=(10,20))
plt.imshow(gt_np24[i][:][:] + c, cmap='gray', vmin=0, vmax=vmax) # 00:00
plt.title(f'True   {str(target)}', fontsize=20)
plt.show()
```

## Multi-step Forecast

```python
table = pd.read_csv('rmse_table_direct1.csv')
table
```

```python
table.rename({'Unnamed: 0':'time_stamp', 'Unnamed: 1':'time_unit', 'Unnamed: 2':'step'},
            axis=1, inplace=True)
```

```python
table
```

```python
table2 = table.T
table2.columns = [f'time_stamp_{i}' for i in table['time_stamp']]
```

```python
table2 = table2.iloc[3:,:]
```

```python
table2
```

```python
x_range = range(2,26,2) 
plt.figure(figsize=(10,7))
plt.plot(x_range, table2['time_stamp_1'], label='0.5H')
plt.plot(x_range, table2['time_stamp_2'], label='1H')
plt.plot(x_range, table2['time_stamp_4'], label='2H')
plt.plot(x_range, table2['time_stamp_6'], label='3H')
plt.plot(x_range, table2['time_stamp_8'], label='4H')
plt.plot(x_range, table2['time_stamp_12'], label='6H')
plt.plot(x_range, table2['time_stamp_16'], label='8H')
plt.plot(x_range, table2['time_stamp_24'], label='12H')
plt.plot(x_range, table2['time_stamp_48'], label='24H')
plt.xticks(x_range)
plt.xlabel('lag', fontsize=15)
plt.ylabel('RMSE', fontsize=15)
plt.title('lag에 따른 RMSE')
plt.legend()
plt.show()
```
