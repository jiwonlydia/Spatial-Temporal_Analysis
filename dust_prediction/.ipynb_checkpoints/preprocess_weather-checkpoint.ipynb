# merge

```python
# 지점, 시작일, 종료일로 merge
# weather['일시'] >= meta['시작일'] & weather['일시'] <= meta['시작일']
```

```python
wea1 = pd.merge(weather, meta, on=['지점','지점명'], how='left')
```

```python
wea1.head()
```

```python
wea1.loc[wea1['종료일'].isna(),'종료일'] = pd.to_datetime('2023-01-01')
```

```python
wea2 = wea1[(wea1['일시'] >= wea1['시작일']) & (wea1['일시'] <= wea1['종료일'])]
wea2.rename({'기온(°C)':'기온'}, axis=1, inplace=True)
wea2.head()
```

```python
wea3 = wea2[['지점명','일시','기온','위도','경도']]
```
