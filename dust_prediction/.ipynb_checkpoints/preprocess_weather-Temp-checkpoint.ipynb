```python
max_lat = meta['위도'].max() + 1 ; print(max_lat)
min_lat = meta['위도'].min() - 1 ; print(min_lat)
unit_lat = (max_lat - min_lat)/10 ; print(unit_lat)
```

```python
max_long = meta['경도'].max() + 1 ; print(max_long)
min_long = meta['경도'].min() - 1; print(min_long)
unit_long = (max_long - min_long)/20 ; print(unit_long)
```

```python
lat_list = []
lat = min_lat
for _ in range(11):
    lat_list.append(lat)
    lat += unit_lat
print(lat_list)

long_list = []
long = min_long
for _ in range(21):
    long_list.append(long)
    long += unit_long
print(long_list)
```

# merge

```python
# 지점, 시작일, 종료일로 merge
# weather['일시'] >= meta['시작일'] & weather['일시'] <= meta['시작일']
```

```python
wea1 = pd.merge(weather, meta, on=['지점','지점명'], how='left')
```

```python
wea1.head()
```

```python
wea1.loc[wea1['종료일'].isna(),'종료일'] = pd.to_datetime('2023-01-01')
```

```python
wea2 = wea1[(wea1['일시'] >= wea1['시작일']) & (wea1['일시'] <= wea1['종료일'])]
wea2.rename({'기온(°C)':'기온'}, axis=1, inplace=True)
```

```python
wea3 = wea2[['지점명','일시','기온','위도','경도']]
wea3
```

```python
print(wea3['지점명'].nunique())
print(wea3['지점명'].unique())
```

```python
wea_tmp = wea3[wea3['일시'] == datetime(2022,7,8,1) + timedelta(hours=0)]
wea_tmp
```

```python
coord = np.zeros(shape=(24,10,20))
coord[0][0][0]
```

```python
print(lat_list)
print(long_list)
```

```python
coord = np.zeros(shape=(24,10,20))
for k in range(24):
    wea_tmp = wea3[wea3['일시'] == (datetime(2022,7,8,1) + timedelta(hours=k))]

    for i in range(10):
        a = (wea_tmp['위도'] >= lat_list[i]) & (wea_tmp['위도'] <= lat_list[i+1])

        for j in range(20):
            b = (wea_tmp['경도'] >= long_list[j]) & (wea_tmp['경도'] <= long_list[j+1])
    #         print('### lat ###', lat_list[i], lat_list[i+1])
    #         print('### long ###', long_list[j], long_list[j+1])

            if len(wea_tmp.loc[a&b,'기온']) > 0:
                val = wea_tmp.loc[a&b,'기온'].mean()
                coord[k][i][j] = val
```

```python
f, ((ax1, ax2, ax3)) = plt.subplots(1, 3, figsize=(10,5))

ax1.imshow(np.log(coord[0][:][:]+1), cmap='gray')
ax2.imshow(np.log(coord[10][:][:]+1), cmap='gray')
ax3.imshow(np.log(coord[20][:][:]+1), cmap='gray')
```

```python
np.save('./data/coord_tmp', coord)
```
