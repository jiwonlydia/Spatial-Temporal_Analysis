```python
max_lat = meta['위도'].max() + 1 ; print(max_lat)
min_lat = meta['위도'].min() - 1 ; print(min_lat)
unit_lat = (max_lat - min_lat)/10 ; print(unit_lat)
```

```python
max_long = meta['경도'].max() + 1 ; print(max_long)
min_long = meta['경도'].min() - 1; print(min_long)
unit_long = (max_long - min_long)/20 ; print(unit_long)
```

```python
lat_list = []
lat = min_lat
for _ in range(11):
    lat_list.append(lat)
    lat += unit_lat
print(lat_list)

long_list = []
long = min_long
for _ in range(21):
    long_list.append(long)
    long += unit_long
print(long_list)
```

# merge

```python
# 지점, 시작일, 종료일로 merge
# weather['일시'] >= meta['시작일'] & weather['일시'] <= meta['시작일']
```

```python
wea1_train = pd.merge(weather_train, meta, on=['지점','지점명'], how='left')
wea1_test = pd.merge(weather_test, meta, on=['지점','지점명'], how='left')
```

```python
display(wea1_train.head(2))
display(wea1_test.head(2))
```

```python
wea1_train.loc[wea1_train['종료일'].isna(),'종료일'] = pd.to_datetime('2023-01-01')
wea1_test.loc[wea1_test['종료일'].isna(),'종료일'] = pd.to_datetime('2023-01-01')
```

```python
wea2_train = wea1_train[(wea1_train['일시'] >= wea1_train['시작일']) & (wea1_train['일시'] <= wea1_train['종료일'])]
wea2_train.rename({'기온(°C)':'기온'}, axis=1, inplace=True)

wea2_test = wea1_test[(wea1_test['일시'] >= wea1_test['시작일']) & (wea1_test['일시'] <= wea1_test['종료일'])]
wea2_test.rename({'기온(°C)':'기온'}, axis=1, inplace=True)
```

```python
wea3_train = wea2_train[['지점명','일시','기온','위도','경도']]
wea3_test = wea2_test[['지점명','일시','기온','위도','경도']]

wea3_train
```

```python
print(wea3_train['지점명'].nunique())
print(wea3_train['지점명'].unique())

print(wea3_test['지점명'].nunique())
# print(wea3_test['지점명'].unique())
```

```python
num_days = 366 # 2020년 2월은 29일까지 있음
num_hours = 24 * num_days

coord_train = np.ones(shape=(num_hours,10,20)) * (-100)

for k in range(num_hours):
    target_datetime = datetime(2020,1,1,0) + timedelta(hours=k)
    print(target_datetime)
    wea_tmp = wea3_train[wea3_train['일시'] == target_datetime]
    
    for i in range(10):
        a = (wea_tmp['위도'] >= lat_list[i]) & (wea_tmp['위도'] <= lat_list[i+1])

        for j in range(20):
            b = (wea_tmp['경도'] >= long_list[j]) & (wea_tmp['경도'] <= long_list[j+1])
            if len(wea_tmp.loc[a&b,'기온']) > 0:
                val = wea_tmp.loc[a&b,'기온'].mean()
                coord_train[k][i][j] = val
```

```python
num_days = 365
num_hours = 24 * num_days

coord_test = np.ones(shape=(num_hours,10,20)) * (-100)

for k in range(num_hours):
    target_datetime = datetime(2021,1,1,0) + timedelta(hours=k)
    print(target_datetime)
    wea_tmp = wea3_test[wea3_test['일시'] == target_datetime]
    
    for i in range(10):
        a = (wea_tmp['위도'] >= lat_list[i]) & (wea_tmp['위도'] <= lat_list[i+1])

        for j in range(20):
            b = (wea_tmp['경도'] >= long_list[j]) & (wea_tmp['경도'] <= long_list[j+1])
            if len(wea_tmp.loc[a&b,'기온']) > 0:
                val = wea_tmp.loc[a&b,'기온'].mean()
                coord_test[k][i][j] = val
```
